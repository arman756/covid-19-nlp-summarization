### Extractive Summarization using NLP Spacy library:

```python
class ExtractiveSummarization():
    
    def __init__(self, text, limit):
        self.text = text
        self.limit = limit
        
        pass
    
    # text — The input text. Can be a short paragraph or a big chuck of text.
    # limit — The number of sentences to be returned.
    def summarize(self):
        print('Summerize the report in the extractive way...')
        # tokenize the input text and find out the important keywords
        keyword = []
        pos_tag = ['PROPN', 'ADJ', 'NOUN', 'VERB']

        #1 — Convert the input text to lower case and tokenize it with spaCy’s language model.
        doc = nlp(self.text.lower())

        #2 — Loop over each of the tokens.    
        for token in doc:

            #3 — Ignore the token if it is a stopword or punctuation.        
            if(token.text in nlp.Defaults.stop_words or token.text in punctuation):
                continue

            #4 — Append the token to a list if it is the part-of-speech tag that we have defined.            
            if(token.pos_ in pos_tag):
                keyword.append(token.text)

        # normalize the weightage of the keywords
        # Counter will convert the list into a dictionary with their respective frequency values.
        freq_word = Counter(keyword)

        # Get the frequency of the top most-common keyword.
        max_freq = Counter(keyword).most_common(1)[0][1]

        # Loop over each item in the dictionary and normalize the frequency. The top most-common keyword will have frequency value of 1.
        for w in freq_word:
            freq_word[w] = (freq_word[w]/max_freq)


        # calculate the importance of the sentences by identifying the occurrence of important keywords and sum up the value.
        sent_strength={}

        # Loop over each sentence in the text. The sentences are split by the spaCy model based on full-stop punctuation.
        for sent in doc.sents:

            # Loop over each word in a sentence based on spaCy’s tokenization.
            for word in sent:

                # Determine if the word is a keyword based on the keywords that we extracted earlier.
                if word.text in freq_word.keys():

                    # Add the normalized keyword value to the key-value pair of the sentence.
                    # Create a new key-value in the sent_strength dictionary using the sentence as key and the normalized keyword value as value.
                    if sent in sent_strength.keys():
                        sent_strength[sent]+=freq_word[word.text]
                    else:
                        sent_strength[sent]=freq_word[word.text]

            # You should be able to get a cumulative normalized value for each sentence. We will use this value to determine the top sentences.

        ### Summerization ###
        summary = []

        # Sort the dictionary based on the normalized value. Set the reverse parameter to True for descending order.
        sorted_x = sorted(sent_strength.items(), key=lambda kv: kv[1], reverse=True)

        # Loop over each of the sorted items.
        counter = 0
        for i in range(len(sorted_x)):

            # Append the results to a list. The first letter of the sentence is capitalized
            summary.append(str(sorted_x[i][0]).capitalize())

            # Break out of the loop if the counter exceeds the limit that we have set. 
            # This determines how many sentences are to be returned from the function.
            counter += 1
            if(counter >= self.limit):
                break

        # Return the list as a string by joining each element with a space.
        return ' '.join(summary)


    def generateReport(self, summerizedText):
            file_path = r'../reports/extractive-summary.txt'
            open(file_path, 'w').close() # clear file first.

            with open(file_path, 'a+') as fp:
                summerizedTextList =  summerizedText.split("   ")
                fp.write('Extractive Summarization:')
                fp.write('\n\n')
                for item in summerizedTextList:
                    fp.write(item)
                    fp.write('\n')

            print('Extractive Summary file is saved!')
            


# Read generated report file and store it in 'text' variable.
text = ''''''
with open('../reports/generated-report.txt','r') as f:
    f.seek(0)
    text = f.read()

    compressText = text.replace('\n',' ')
    text = compressText

print('Generated report file has been read.')
        
extractiveSummarization = ExtractiveSummarization(text, 3)
summerizedExtractiveText = extractiveSummarization.summarize()
extractiveSummarization.generateReport(summerizedExtractiveText)

summerizedExtractiveText
```

### Abstractive Summarization using NLP Spacy library:

We’ll use SpaCy to import a pre-trained NLP pipeline to help interpret the grammatical structure of the text.

```python
class AbstractiveSummarization():
    
    def __init__(self, text, per):
        self.text = text
        self.per = per
        
        pass
    
    # per is the percentage (0 to 1) of sentences you want to extract.
    def summarize(self):
        print('Summerize the report in the abstractive way...')
        # Tokenize the text with the SpaCy pipeline.
        doc= nlp(self.text)
        tokens=[token.text for token in doc]
        word_frequencies={}

        # Count the number of times a word is used (not including stop words or punctuation)
        for word in doc:
            if word.text.lower() not in list(STOP_WORDS):
                if word.text.lower() not in punctuation:            
                    if word.text not in word_frequencies.keys():
                        word_frequencies[word.text] = 1
                    else:
                        word_frequencies[word.text] += 1

        # normalize the count. A word that’s used more frequently has a higher normalized count.
        max_frequency=max(word_frequencies.values())
        for word in word_frequencies.keys():
            word_frequencies[word]=word_frequencies[word]/max_frequency

        sentence_tokens= [sent for sent in doc.sents]
        sentence_scores = {}

        # Calculate the sum of the normalized count for each sentence.
        for sent in sentence_tokens:
            for word in sent:
                if word.text.lower() in word_frequencies.keys():
                    if sent not in sentence_scores.keys():                            
                        sentence_scores[sent]=word_frequencies[word.text.lower()]
                    else:
                        sentence_scores[sent]+=word_frequencies[word.text.lower()]

        # Extract a percentage of the highest ranked sentences. These serve as our summary.
        select_length=int(len(sentence_tokens)*self.per)
        summary=nlargest(select_length, sentence_scores,key=sentence_scores.get)
        final_summary=[word.text for word in summary]
        summary=''.join(final_summary)
        return summary

    def generateReport(self, summerizedText):   
        file_path = r'../reports/abstractive-summary.txt'
        open(file_path, 'w').close() # clear file first.

        with open(file_path, 'a+') as fp:
            summerizedTextList =  summerizedText.split("   ")
            fp.write('Abstractive Summarization:')
            fp.write('\n\n')
            for item in summerizedTextList:
                fp.write(item)
                fp.write('\n')

        print('Abstractive Summary file is saved!')



# Read generated report file and store it in 'text' variable.
text = ''''''
with open('../reports/generated-report.txt','r') as f:
    f.seek(0)
    text = f.read()

    compressText = text.replace('\n',' ')
    text = compressText

print('Generated report file has been read.')

# per is the percentage (0 to 1) of sentences you want to extract.
abstractiveSummarization = AbstractiveSummarization(text, 0.05)
summerizedAbstractiveText = abstractiveSummarization.summarize()
abstractiveSummarization.generateReport(summerizedAbstractiveText)

summerizedAbstractiveText
```
